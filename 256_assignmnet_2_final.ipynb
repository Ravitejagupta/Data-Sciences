```python
import numpy as np
import pandas as pd
from sklearn.metrics import pairwise
```

```python
header = ['ID', 'Rating [Rogue One/Star Wars]','Rating [Fight Club]','Rating [The Lord of the Rings]','Rating [Trolls]','Rating [Despicable Me]','Rating [Kubo]','Rating [The hangover]','Rating [Captain America]','Rating [The big Lebowski]','Rating [Almost famous]','Rating [The hunger games]','Rating [Pulp Fiction]','Rating [Sing ]','Rating [500 days of Summer]','Rating [Twilight]','Rating [Lala Land]']
df_train = pd.read_csv('Desktop/train.csv')
df_train = df_train[header]
df_train = df_train.fillna(0)
```

```python
df_test = pd.read_csv('Desktop/test.csv')
df_test = df_test[header]
df_test = df_test.fillna(0)

df_test
```

```python
train = pd.DataFrame(df_train.iloc[1:]).values
test = pd.DataFrame(df_test.iloc[1:]).values

test
```

```python
cosine_user_sim = pd.DataFrame(pairwise.pairwise_distances(train, test, metric = 'cosine'))
t = pd.concat([cosine_user_sim, df_train.loc[:,'Rating [Despicable Me]']], axis = 1)
valid_data = t.loc[t['Rating [Despicable Me]'] != 0]
valid_data
```

```python
final = []
for j in range(len(test)):
    to_be_considered = t.iloc[:,[j,52]].sort_values(by = j).iloc[-20:,:]
    predicted_value = (to_be_considered.iloc[:,0] * to_be_considered.iloc[:,1]).sum() / to_be_considered.iloc[:,0].sum()
    final.append(predicted_value)

final = [round(int(i)) for i in final]

final
```

```python
predictions = pd.concat([df_test.ID[:-1], pd.Series(final)], axis = 1)

predictions
```

```python
predictions.to_csv('Desktop/Output_Assignment_2_final.csv')
```
