```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

reader = pd.read_excel
path = 'Downloads/ML_Iris.xlsx'
data = reader(path)
```

```python
plt.figure(figsize= (12,8))
plt.plot(data["Column1"])
plt.plot(data["Column2"])
plt.plot(data["Column3"])
plt.plot(data["Column4"])
plt.show()
```

```python
m1 = data["Column1"].mean()
m2 = data["Column2"].mean()
m3 = data["Column3"].mean()
m4 = data["Column4"].mean()

print(m1,m2,m3,m4)
```

```python
v1 = data["Column1"].var()
v2 = data["Column2"].var()
v3 = data["Column3"].var()
v4 = data["Column4"].var()

print(v1,v2,v3,v4)
```

```python
me1 = data["Column1"].median()
me2 = data["Column2"].median()
me3 = data["Column3"].median()
me4 = data["Column4"].median()

print(me1,me2,me3,me4)
```
