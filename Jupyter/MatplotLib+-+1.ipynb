```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
%matplotlib inline
```

```python
 x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x,y,'r')
plt.xlabel('x')
plt.ylabel('squares')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([1,1,.5,.5])
print(axes)
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1,.1,.8,.8])
axes1.set_title("Larger Plot")
axes1.plot(x,y)

axes2 = fig.add_axes([.2,.5,.4,.3])
axes2.set_title("Smaller Plot")
axes2.plot(y,x)
```
